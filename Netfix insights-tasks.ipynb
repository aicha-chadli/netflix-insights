```python
#Importer les bibliothèques de visualisation
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Lecture du dataset
df = pd.read_csv("C:/Users/hp/Downloads/netflix_titles.csv")

# Style des graphiques
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
#a. Répartition du type d'œuvres (films vs séries)
sns.countplot(x="type", data=df, palette="pastel")
plt.title("Répartition du type d'œuvres (Films vs Séries)")
plt.xlabel("Type")
plt.ylabel("Nombre d'œuvres")
plt.show()
```

```python
#b. Répartition des œuvres par pays
top_countries = df['country'].value_counts().nlargest(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette="muted")
plt.title("Top 10 des pays producteurs")
plt.xlabel("Nombre d'œuvres")
plt.ylabel("Pays")
plt.show()
```

```python
#c. Répartition par année de sortie
sns.histplot(df['release_year'], bins=30, kde=False, color="skyblue")
plt.title("Répartition des années de sortie")
plt.xlabel("Année")
plt.ylabel("Nombre d'œuvres")
plt.show()
```

```python
 #d. Répartition des ratings
rating_counts = df['rating'].value_counts().sort_values(ascending=False)
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette="coolwarm")
plt.title("Répartition des ratings")
plt.xlabel("Rating")
plt.ylabel("Nombre d'œuvres")
plt.xticks(rotation=45)
plt.show()
```

```python
#g. Répartition des genres
# On isole les films et on extrait la durée numérique
movies_df = df[df["type"] == "Movie"].copy()
movies_df["duration_minutes"] = movies_df["duration"].str.extract("(\d+)").astype(float)

# Histogramme
sns.histplot(movies_df["duration_minutes"].dropna(), bins=30, kde=True, color="salmon")
plt.title("Répartition de la durée des films (en minutes)")
plt.xlabel("Durée")
plt.ylabel("Nombre de films")
plt.show()


```

```python
#Répartition de la durée des séries
# Séries uniquement + extraction du nombre de saisons
series_df = df[df["type"] == "TV Show"].copy()
series_df["duration_seasons"] = series_df["duration"].str.extract("(\d+)").astype(float)

sns.histplot(series_df["duration_seasons"].dropna(), bins=20, kde=False, color="lightgreen")
plt.title("Répartition de la durée des séries (en saisons)")
plt.xlabel("Nombre de saisons")
plt.ylabel("Nombre de séries")
plt.show()
```

```python
#Répartition des genres d’œuvres
# Séparer les genres (listed_in) et compter les occurrences
genre_series = df["listed_in"].dropna().str.split(", ")
all_genres = genre_series.explode()
top_genres = all_genres.value_counts().nlargest(15)

sns.barplot(x=top_genres.values, y=top_genres.index, palette="magma")
plt.title("Top 15 des genres les plus présents")
plt.xlabel("Nombre d’œuvres")
plt.ylabel("Genres")
plt.show()
```

```python
#Top 5 des séries les plus longues
top_series = series_df.sort_values("duration_seasons", ascending=False).dropna(subset=["duration_seasons"])
top_series[["title", "duration", "duration_seasons"]].head(5)
```

```python
#Top 5 des films les plus longs
top_movies = movies_df.sort_values("duration_minutes", ascending=False).dropna(subset=["duration_minutes"])
top_movies[["title", "duration", "duration_minutes"]].head(5)
```

```python
#Répartition des réalisateurs des œuvres françaises
# Filtrer les œuvres dont le pays contient "France"
french_df = df[df["country"].fillna("").str.contains("France")]
top_directors = french_df["director"].dropna().value_counts().nlargest(10)

sns.barplot(x=top_directors.values, y=top_directors.index, palette="viridis")
plt.title("Top réalisateurs d’œuvres françaises")
plt.xlabel("Nombre d’œuvres")
plt.ylabel("Réalisateurs")
plt.show()
```

```python
#Répartition des œuvres par date d’ajout au catalogue
# Conversion de la date
df["date_added"] = pd.to_datetime(df["date_added"], errors='coerce')

# Regrouper par année
df["year_added"] = df["date_added"].dt.year
yearly_adds = df["year_added"].value_counts().sort_index()

sns.lineplot(x=yearly_adds.index, y=yearly_adds.values, marker="o", color="teal")
plt.title("Répartition des œuvres par année d’ajout")
plt.xlabel("Année")
plt.ylabel("Nombre d’ajouts")
plt.show()
```

```python
#Graphique libre – Exemple : Évolution des ajouts Films vs Séries
# Regrouper par type et année d'ajout
df_grouped = df.groupby(["year_added", "type"]).size().unstack()

df_grouped.plot(kind="bar", stacked=True, colormap="Accent", figsize=(12,6))
plt.title("Ajouts annuels : Films vs Séries")
plt.xlabel("Année")
plt.ylabel("Nombre d’œuvres ajoutées")
plt.xticks(rotation=45)
plt.legend(title="Type")
plt.tight_layout()
plt.show()
```
